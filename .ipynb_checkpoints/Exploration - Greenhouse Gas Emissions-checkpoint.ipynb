```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
%matplotlib inline
```

```python
# Read Greenhouse Data
greenhouse_gas_data = pd.read_csv('Datasets/CO2EmissionsData.csv')
greenhouse_gas_data
```

```python
print(greenhouse_gas_data.columns)
greenhouse_gas_data.head(2) # get first two rows
```

```python
greenhouse_gas_data.info() # get column types
```

```python
len(greenhouse_gas_data) # get number of rows
```

```python
# Rename countries to camelcase
greenhouse_gas_data['Country'] = greenhouse_gas_data['Country'].str.title()
```

```python
# Get unique values for countries
greenhouse_gas_data['Country'].unique()
```

```python
len(greenhouse_gas_data['Country'].unique()) # get number of countries
```

```python
# Get country with most emissions
gas_most_by_country_data = greenhouse_gas_data.groupby([greenhouse_gas_data['Country'], 'Year'])[['Total']].max()
gas_most_by_country_data.columns = ['Emissions']
gas_most_by_country_data = gas_most_by_country_data.round(2)
gas_most_by_country_data = gas_most_by_country_data.sort_values('Emissions', ascending=False).reset_index()
gas_most_by_country_data.head(5)
```

```python
# Get country with least emissions
gas_least_by_country_data = greenhouse_gas_data.groupby([greenhouse_gas_data['Country'], 'Year'])[['Total']].min()
gas_least_by_country_data.columns = ['Emissions']
gas_least_by_country_data = gas_least_by_country_data.round(2)
gas_least_by_country_data = gas_least_by_country_data.sort_values('Emissions', ascending=True).reset_index()
gas_least_by_country_data.head(5)
```

```python
# Get a subsample with US data
gas_us_data = greenhouse_gas_data[greenhouse_gas_data['Country']=='United States Of America']
gas_us_data
```

```python
gas_us_data.set_index('Year').plot(figsize=(10,5), grid=True)
```

```python
gas_us_data.plot(kind='scatter', x='Year', y='Total', figsize=(10,5), grid=True)
```

```python
# Group by Country to get the an average
gas_grouped_by_country_data = greenhouse_gas_data.groupby([greenhouse_gas_data['Country']])[['Total']].mean()
gas_grouped_by_country_data = gas_grouped_by_country_data.round(2)
gas_grouped_by_country_data = gas_grouped_by_country_data.reset_index()
gas_grouped_by_country_data = gas_grouped_by_country_data[gas_grouped_by_country_data['Total']>5000]
gas_grouped_by_country_data
```

```python
names = gas_grouped_by_country_data.values.T.tolist()[0]
values = gas_grouped_by_country_data.values.T.tolist()[1]

#fig, axs = plt.subplots(1, 3, figsize=(20, 5), sharey=True)
#axs[0].bar(names, values)
#axs[1].scatter(names, values)
#axs[2].plot(names, values)
fig, axs = plt.subplots(figsize=(20, 7))
axs.bar(names, values)
plt.xticks(rotation=90)
fig.suptitle('Categorical Plotting')
```

```python
# Group by Country to get the an average for the countries with most emissions
gas_grouped_by_country_data_pie = greenhouse_gas_data.groupby([greenhouse_gas_data['Country']])[['Total']].mean()
gas_grouped_by_country_data_pie = gas_grouped_by_country_data.round(2)
gas_grouped_by_country_data_pie = gas_grouped_by_country_data.reset_index()
gas_grouped_by_country_data_pie = gas_grouped_by_country_data[gas_grouped_by_country_data['Total']>50000]

names = gas_grouped_by_country_data_pie.values.T.tolist()[0]
values = gas_grouped_by_country_data_pie.values.T.tolist()[1]
 
# Plot
plt.pie(values, labels=names, autopct='%1.1f%%', shadow=True, startangle=90)
plt.axis('equal')
plt.show()

#patches, texts = plt.pie(values, shadow=True, startangle=90)
#plt.legend(patches, names, loc="best")
#plt.axis('equal')
#plt.tight_layout()
#plt.show()
```

```python
### Get emissions trend through the years
gas_grouped_by_year_data = greenhouse_gas_data.groupby([greenhouse_gas_data['Year']])[['Total']].sum()
gas_grouped_by_year_data = gas_grouped_by_year_data.round(2)
gas_grouped_by_year_data = gas_grouped_by_year_data.reset_index()
gas_grouped_by_year_data
```

```python
gas_grouped_by_year_data.set_index('Year').plot(figsize=(10,5), grid=True)
```

```python
# Get average emissions trend through the years
gas_grouped_by_year_avg_data = greenhouse_gas_data.groupby([greenhouse_gas_data['Year']])[['Total']].mean()
gas_grouped_by_year_avg_data = gas_grouped_by_year_avg_data.round(2)
gas_grouped_by_year_avg_data = gas_grouped_by_year_avg_data.reset_index()
gas_grouped_by_year_avg_data
```

```python
gas_grouped_by_year_avg_data.set_index('Year').plot(figsize=(10,5), grid=True)
```
