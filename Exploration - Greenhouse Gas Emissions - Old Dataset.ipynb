```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
%matplotlib inline
```

```python
# Read Greenhouse Data
greenhouse_gas_data = pd.read_csv('Datasets/GreenhouseGasInventoryData.csv')
greenhouse_gas_data
```

```python
print(greenhouse_gas_data.columns)
greenhouse_gas_data.head(2) # get first two rows
```

```python
greenhouse_gas_data.info() # get column types
```

```python
len(greenhouse_gas_data) # get number of rows
```

```python
# Get unique values for countries
greenhouse_gas_data['country_or_area'].unique()
```

```python
len(greenhouse_gas_data['country_or_area'].unique()) # get number of countries
```

```python
# Get only the category we are looking for - greenhouse gas without LUCUFU
gas_only_data = greenhouse_gas_data[greenhouse_gas_data['category']=='greenhouse_gas_ghgs_emissions_without_land_use_land_use_change_and_forestry_lulucf_in_kilotonne_co2_equivalent']
gas_only_data = gas_only_data.round(2) # round value to 2 decimals
gas_only_data
```

```python
# Get country with most emissions
gas_most_by_country_data = gas_only_data.groupby([gas_only_data['country_or_area'], 'year'])[['value']].max()
gas_most_by_country_data.columns = ['Emissions']
gas_most_by_country_data = gas_most_by_country_data.round(2)
gas_most_by_country_data = gas_most_by_country_data.sort_values('Emissions', ascending=False).reset_index()
gas_most_by_country_data.head(5)
```

```python
# Get country with least emissions
gas_least_by_country_data = gas_only_data.groupby([gas_only_data['country_or_area'], 'year'])[['value']].min()
gas_least_by_country_data.columns = ['Emissions']
gas_least_by_country_data = gas_least_by_country_data.round(2)
gas_least_by_country_data = gas_least_by_country_data.sort_values('Emissions', ascending=True).reset_index()
gas_least_by_country_data.head(5)
```

```python
# Get a subsample with US data
gas_us_data = gas_only_data[gas_only_data['country_or_area']=='United States of America']
gas_us_data
```

```python
gas_us_data.set_index('year').plot(figsize=(10,5), grid=True)
```

```python
gas_us_data.plot(kind='scatter', x='year', y='value', figsize=(10,5), grid=True)
```

```python
# Group by Country to get the an average
gas_grouped_by_country_data = gas_only_data.groupby([gas_only_data['country_or_area']])[['value']].mean()
gas_grouped_by_country_data = gas_grouped_by_country_data.round(2)
gas_grouped_by_country_data = gas_grouped_by_country_data.reset_index()
gas_grouped_by_country_data
```

```python
names = gas_grouped_by_country_data.values.T.tolist()[0]
values = gas_grouped_by_country_data.values.T.tolist()[1]

#fig, axs = plt.subplots(1, 3, figsize=(20, 5), sharey=True)
#axs[0].bar(names, values)
#axs[1].scatter(names, values)
#axs[2].plot(names, values)
fig, axs = plt.subplots(figsize=(20, 7))
axs.bar(names, values)
plt.xticks(rotation=90)
fig.suptitle('Categorical Plotting')
```

```python
# Get emissions trend through the years
gas_grouped_by_year_data = gas_only_data.groupby([gas_only_data['year']])[['value']].sum()
gas_grouped_by_year_data = gas_grouped_by_year_data.round(2)
gas_grouped_by_year_data = gas_grouped_by_year_data.reset_index()
gas_grouped_by_year_data
```

```python
gas_grouped_by_year_data.set_index('year').plot(figsize=(10,5), grid=True)
```

```python
# Get average emissions trend through the years
gas_grouped_by_year_avg_data = gas_only_data.groupby([gas_only_data['year']])[['value']].mean()
gas_grouped_by_year_avg_data = gas_grouped_by_year_avg_data.round(2)
gas_grouped_by_year_avg_data = gas_grouped_by_year_avg_data.reset_index()
gas_grouped_by_year_avg_data
```

```python
gas_grouped_by_year_avg_data.set_index('year').plot(figsize=(10,5), grid=True)
```
