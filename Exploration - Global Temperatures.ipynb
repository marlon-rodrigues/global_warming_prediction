```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
%matplotlib inline
```

```python
# Read Global Temps by Country
global_temps_by_country_data = pd.read_csv('Datasets/GlobalLandTemperaturesByCountry.csv')
global_temps_by_country_data
```

```python
print(global_temps_by_country_data.columns)
global_temps_by_country_data.head(2) # get first two rows
```

```python
global_temps_by_country_data.info() # get column types
```

```python
len(global_temps_by_country_data) # get number of rows
```

```python
# Get unique values for countries
global_temps_by_country_data['Country'].unique()
```

```python
len(global_temps_by_country_data['Country'].unique()) # get number of countries
```

```python
len(global_temps_by_country_data['dt'].unique()) # get number of dates
```

```python
# Group by country to check the number of available data for each country
grouped_by_country = global_temps_by_country_data.groupby('Country')[['dt']].count()
grouped_by_country.columns = ['Number of Metrics']
sorted_group = grouped_by_country.sort_values('Number of Metrics', ascending=True).reset_index()
sorted_group
```

```python
# Get max temperature by country
max_temp_group = global_temps_by_country_data.groupby('Country')[['AverageTemperature']].max()
max_temp_group.columns = ['Max Temp']
sorted_max_temp_group = max_temp_group.sort_values('Max Temp', ascending=False).reset_index()
sorted_max_temp_group.head(5)
```

```python
# Get min temperature by country
min_temp_group = global_temps_by_country_data.groupby(['Country'], sort=False)[['AverageTemperature']].min()
min_temp_group.columns = ['Min Temp']
sorted_max_temp_group = min_temp_group.sort_values('Min Temp', ascending=True).reset_index()
sorted_max_temp_group.head(5)
```

```python
# Get average temperature by country
avg_temp_group = global_temps_by_country_data.groupby(['Country'], sort=False)[['AverageTemperature']].mean()
avg_temp_group.columns = ['Avg Temp']
sorted_avg_temp_group = avg_temp_group.sort_values('Avg Temp', ascending=True).reset_index()
sorted_avg_temp_group
```

```python
# Get average temp for US only
US_avg_temp = sorted_avg_temp_group[sorted_avg_temp_group['Country']=='United States']
US_avg_temp
```

```python
# Get all temperatures for US only
US_temps = global_temps_by_country_data[global_temps_by_country_data['Country']=='United States']
US_temps
```

```python
# Convert date column to date-time 
temps_period = global_temps_by_country_data
temps_period['dt'] = pd.to_datetime(global_temps_by_country_data['dt'])
temps_period
```

```python
temps_period.info()
```

```python
# Get total number of rows missing a value
temps_period.isna().sum()
```

```python
# Convert NaN to zero (0)
temps_period.fillna(0, inplace=True)
temps_period
```

```python
# Average temperature by year in the US
temps_data_us = temps_period[temps_period['Country']=='United States']

temps_data_us_avg_temp = temps_data_us.groupby(temps_data_us['dt'].dt.year)[['AverageTemperature']].mean()
temps_data_us_avg_temp
```

```python
# Get average temp with country column
temps_data_us_avg_temp_with_country = temps_data_us.groupby([temps_data_us['dt'].dt.year, 'Country'])[['AverageTemperature']].mean()
#temps_data_us_avg_temp_with_country.columns = ['Avg Temp']
temps_data_us_avg_temp_with_country.tail(5)
```

```python
# Filter data for the 1990 - 2013 period
type(temps_data_us_avg_temp_with_country)
temps_data_us_avg_temp_with_country.info()
```

```python
temps_data_us_avg_temp_with_country = temps_data_us_avg_temp_with_country.reset_index()
#temps_data_us_avg_temp_with_country = temps_data_us_avg_temp_with_country.set_index(temps_data_us_avg_temp_with_country['dt'], inplace=True)
temp_sub_period = temps_data_us_avg_temp_with_country[temps_data_us_avg_temp_with_country['dt']>1989]
#temp_sub_period = temps_data_us_avg_temp_with_country[temps_data_us_avg_temp_with_country.index.get_level_values('dt')>1989]
temp_sub_period.info()
temp_sub_period
```
